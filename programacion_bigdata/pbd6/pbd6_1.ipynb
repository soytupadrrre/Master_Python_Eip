## DASK <a class="anchor" id="dask"></a>

```python
%%time
import dask.dataframe as dd
```

```python
%%time
df_dask = dd.read_csv('data/yellow_tripdata_2018-12.csv', assume_missing=True)
df_dask.head()
```

```python
%%time
df_dask.describe()
```

```python
%%time
df_dask.describe().compute()
```

```python
%%time
df_dask.fare_amount.value_counts()
```

```python
%%time
df_dask.fare_amount.value_counts().compute()
```
