# Script Twitter Streaming API
Creación de script que permite extraer tweets de twitter filtrados, posteriormente se envían estos scripts al cuaderno 1 para recibirlos con Spark Streaming

```python
import socket
import sys
# pip install tweepy
import tweepy
import json
# File with twitter API secrets
import twitter_secrets as ts
import datetime
```

Creación de clase para poder manejar los tweets recibidos

```python
class TweetStreamListener(tweepy.StreamingClient):
    TCP_IP = "localhost"
    TCP_PORT = 9999
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.bind((TCP_IP, TCP_PORT))
    s.listen(1)
    print("Waiting for TCP connection...")
    conn, addr = s.accept()
    print("Connected... Starting getting tweets.")
    
    def send_tweets_to_spark(self, tweet, tcp_connection):
        tweet += '\n'
        tcp_connection.send(tweet.encode())

    def on_data(self, data):
        data = json.loads(data)["data"]
        date = datetime.date.today().strftime("%Y-%m-%d")
        
        try: 
            text:str = data["extended_tweet"]["full_text"]
        except:
            text:str = data["text"]

        text = text.replace('\n', '')
        text = text.replace('\t', '')

        tweet = f"{date}\t{text}"
        print(tweet)
        print("-------------------------------")
        self.send_tweets_to_spark(tweet, self.conn)

        return True

    def on_error(self, status):
        print(status)
```

Lanzamiento de Twitter Stream API

```python
stream = TweetStreamListener(ts.BEARER_TOKEN)
stream.add_rules(tweepy.StreamRule("Trump"))
stream.filter()
```
